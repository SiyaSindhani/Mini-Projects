```python
#Problem Statement : Blanks with operator 6_5_4_3_2 = 25
```

```python
import random as r             #Library Inclusion
```

```python
#Parameter Setting
seq     = ['6','5','4','3','2']
op      = ['+','-','*','/']
Target  = 25
# Number of Inerations
Iterations  = 10000    
```

```python
finalSeq = []
finalSeq.append(seq[0])
print(finalSeq)

for i in seq[1:]:
  finalSeq.append(''.join(r.sample(op,1)))
  finalSeq.append(i)
  print(finalSeq)

exp=''.join(finalSeq)
print("\nExpression: ", exp)

print("Expression Value: ", eval(exp))
```

```python
# Loop till number of Iterations

f=0
for i in range(Iterations):

  finalSeq = []
  finalSeq.append(seq[0])
  for i in seq[1:]:
    finalSeq.append(''.join(r.sample(op,1)))
    finalSeq.append(i)
  
  exp=''.join(finalSeq)
  print (exp, "  ", eval(exp))

  if eval(exp) == Target:
    f=1
    break

    
if f==0:
  print("No Sequence found !!")
```
